```python
#convert date into datetime
data_btc['Date'] = data_btc['Date'].astype('datetime64[ns]')
data_eth['Date'] = data_eth['Date'].astype('datetime64[ns]')
data_ltc['Date'] = data_ltc['Date'].astype('datetime64[ns]')
data_xmr['Date'] = data_xmr['Date'].astype('datetime64[ns]')
data_xrp['Date'] = data_xrp['Date'].astype('datetime64[ns]')
data_sol['Date'] = data_sol['Date'].astype('datetime64[ns]')
data_xlm['Date'] = data_xlm['Date'].astype('datetime64[ns]')
data_usdt['Date'] = data_usdt['Date'].astype('datetime64[ns]')
data_trx['Date'] = data_trx['Date'].astype('datetime64[ns]')
```

```python
#correcting date format
data_btc['Date'] = data_btc['Date'].dt.strftime('%d-%m-%Y')
data_eth['Date'] = data_eth['Date'].dt.strftime('%d-%m-%Y')
data_ltc['Date'] = data_ltc['Date'].dt.strftime('%d-%m-%Y')
data_xmr['Date'] = data_xmr['Date'].dt.strftime('%d-%m-%Y')
data_xrp['Date'] = data_xrp['Date'].dt.strftime('%d-%m-%Y')
data_sol['Date'] = data_sol['Date'].dt.strftime('%d-%m-%Y')
data_xlm['Date'] = data_xlm['Date'].dt.strftime('%d-%m-%Y')
data_usdt['Date'] = data_usdt['Date'].dt.strftime('%d-%m-%Y')
data_trx['Date'] = data_trx['Date'].dt.strftime('%d-%m-%Y')
```

```python
data1_btc = data_btc.set_index(pd.DatetimeIndex(data_btc['Date'].values))
data2_eth = data_eth.set_index(pd.DatetimeIndex(data_eth['Date'].values))
data3_ltc = data_ltc.set_index(pd.DatetimeIndex(data_ltc['Date'].values))
data4_xmr = data_xmr.set_index(pd.DatetimeIndex(data_xmr['Date'].values))
data5_xrp = data_xrp.set_index(pd.DatetimeIndex(data_xrp['Date'].values))
data6_sol = data_sol.set_index(pd.DatetimeIndex(data_sol['Date'].values))
data7_xlm = data_xlm.set_index(pd.DatetimeIndex(data_xlm['Date'].values))
data8_usdt = data_usdt.set_index(pd.DatetimeIndex(data_usdt['Date'].values))
data9_trx = data_trx.set_index(pd.DatetimeIndex(data_trx['Date'].values))
```

```python
data_btc = data_btc.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_eth = data_eth.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_ltc = data_ltc.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_xmr = data_xmr.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_xrp = data_xrp.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_sol = data_sol.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_usdt = data_usdt.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
data_trx = data_trx.query("Date >= '2020-04-01' \
                       and Date < '2021-04-01'")
```
